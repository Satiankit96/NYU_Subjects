## 8. Show the total number of patients with normal blood pressure (systolic blood pressure less than 120 mm Hg).

## 9. Show the percentage of females with high blood pressure (SBP of 130 mm Hg or higher).

```python
# all female records
dataFemale = data[(data["Female"]==True)]

# finding number of female records 
totalFemales = len(dataFemale)
highBpFemales = len(dataFemale[(dataFemale["SBP"]>=130)])

# percentage
percFemaleHighBp = (highBpFemales/totalFemales)*100
```

```python
print("Percentage of Females with High Blood Pressure:",percFemaleHighBp)
```

## 10. Build a bar chart of the number of patients by the blood sugar levels (normal, prediabetes, and diabetes) at baseline.
- A normal A1C level is below 5.7%. 
- A level of 5.7% to 6.4% indicates prediabetes. 
- A level of 6.5% or more indicates diabetes.

```python
# calculating length of records
normalCount = len(data[(data["HemoglobinA1C.Baseline"] < 5.7)])
prediabetesCount = len(data[(data["HemoglobinA1C.Baseline"] >= 5.7) & (data["HemoglobinA1C.Baseline"] <= 6.4)])
diabetesCount = len(data[(data["HemoglobinA1C.Baseline"] >= 6.5)])
# creating list of record lengths
lenData = [normalCount, prediabetesCount, diabetesCount]
```

```python
plt.bar(["Normal","Prediabetes","Diabetes"],lenData, width=0.5, color="red")
plt.title("Barplot of Patients by Blood Sugar Levels")
plt.show()
```
